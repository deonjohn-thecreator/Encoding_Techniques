```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Create dummy hospital dataset

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'AppointmentType': ['General', 'Dental', 'Cardiology', 'General', 'Dental'],
    'City': ['Kochi', 'Kottayam', 'Calicut', 'Kochi', 'Calicut']
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display dataset
print(df)
```

```python
# Display dataset information
print(df.info())
```

```python
# Statistical summary
print(df.describe())
```

```python
# Create copy of dataset
label_encoded_df = df.copy()

# Initialize LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to each column
label_encoded_df['Gender'] = le.fit_transform(label_encoded_df['Gender'])

label_encoded_df['AppointmentType'] = le.fit_transform(label_encoded_df['AppointmentType'])

label_encoded_df['City'] = le.fit_transform(label_encoded_df['City'])

# Display encoded dataset
print(label_encoded_df)
```

## Observation of Label Encoding

Label Encoding converts categorical values into numerical labels.

Example:
- Male → 1
- Female → 0

Advantages:
- Simple and memory efficient
- Good for ordinal data

Disadvantages:
- Machine learning models may assume numerical order
- Not suitable for nominal categorical data

```python
# Apply One-Hot Encoding
one_hot_encoded_df = pd.get_dummies(df)

# Display encoded dataset
print(one_hot_encoded_df)
```

## Observation of One-Hot Encoding

One-Hot Encoding creates separate binary columns for each category.

Example:
- Gender_Male
- Gender_Female

Advantages:
- No false numerical relationship
- Better for nominal categorical data

Disadvantages:
- Increases number of columns
- Uses more memory

```python
# Compare shapes

print("Original Dataset Shape:", df.shape)

print("Label Encoded Shape:", label_encoded_df.shape)

print("One-Hot Encoded Shape:", one_hot_encoded_df.shape)
```

# Comparative Analysis

## Label Encoding
- Converts categories into numeric labels
- Dataset size remains same
- Suitable for ordinal data
- Can create misleading relationships

## One-Hot Encoding
- Creates binary columns
- Dataset expands in size
- Suitable for nominal data
- Avoids false ordering problem

## Conclusion
One-Hot Encoding is generally preferred for nominal categorical variables, while Label Encoding is useful for ordinal categories.
